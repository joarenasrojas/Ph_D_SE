```python
np.random.binomial(10, 0.5, x).std()
```

```python
np.random.seed(123)
std_dev_1 = []
media_1 = []
histogramas_1 = []
contador = 0
for x in range(1,10001):
    contador = contador + 1
    media_muestral = []
    for j in range(500):
        muestra = np.random.binomial(10, 0.5, x)
        media_muestral.append(muestra.mean())
    media_muestral = np.asarray(media_muestral)
    if contador > 30:
        std_dev_1.append(media_muestral.std())
        media_1.append(media_muestral.mean())
    if contador in [10,20,30,100, 500,1000, 10000]:
        temp = [contador, media_muestral]
        histogramas_1.append(temp)
# result of flipping a coin 10 times, tested 1000 times.
```

## 2) Probar con distintas distribuciones: Normal, Uniforme, Exponencial ; y con distintos tamaños de muestra (n).

NORMAL

```python
np.random.seed(123)
std_dev_normal = []
media_normal = []
histogramas_normal = []
contador = 0
for x in range(1,10001):
    contador = contador + 1
    media_muestral = []
    for j in range(500):
        muestra = np.random.normal(0,1,x)
        media_muestral.append(muestra.mean())
    media_muestral = np.asarray(media_muestral)
    if contador > 30:
        std_dev_normal.append(media_muestral.std())
        media_normal.append(media_muestral.mean())
    if contador in [10,20,30,100, 500,1000, 10000]:
        temp = [contador, media_muestral]
        histogramas_normal.append(temp)
```

UNIFORME

```python
np.random.seed(123)
std_dev_unif = []
media_unif = []
histogramas_unif = []
contador = 0
for x in range(1,10001):
    contador = contador + 1
    media_muestral = []
    for j in range(500):
        muestra = np.random.uniform(-5,5,x)
        media_muestral.append(muestra.mean())
    media_muestral = np.asarray(media_muestral)
    if contador > 30:
        std_dev_unif.append(media_muestral.std())
        media_unif.append(media_muestral.mean())
    if contador in [10,20,30,100, 500,1000, 10000]:
        temp = [contador, media_muestral]
        histogramas_unif.append(temp)
```

EXPONENCIAL

```python
np.random.seed(123)
std_dev_exp = []
media_exp = []
histogramas_exp = []
contador = 0
for x in range(1,10001):
    contador = contador + 1
    media_muestral = []
    for j in range(500):
        muestra = np.random.exponential(size = x)
        media_muestral.append(muestra.mean())
    media_muestral = np.asarray(media_muestral)
    if contador > 30:
        std_dev_exp.append(media_muestral.std())
        media_exp.append(media_muestral.mean())
    if contador in [10,20,30,100, 500,1000, 10000]:
        temp = [contador, media_muestral]
        histogramas_exp.append(temp)
# result of flipping a coin 10 times, tested 1000 times.
```
